```python
import csv
import shutil
from pathlib import Path

VALID_IMAGE_EXTENSIONS = {".jpg", ".jpeg", ".png", ".bmp", ".webp"}

if MIXED_BUNDLE_ROOT.exists():
    shutil.rmtree(MIXED_BUNDLE_ROOT)

for split_name in ["train", "val"]:
    for class_name in ["open", "closed"]:
        (MIXED_BUNDLE_ROOT / split_name / class_name).mkdir(parents=True, exist_ok=True)

rows = []

def copy_split_sources(clean_root: Path, lowlight_root: Path, split_name: str):
    for class_name in ["open", "closed"]:
        dst_dir = MIXED_BUNDLE_ROOT / split_name / class_name

        for domain_name, src_root in [("clean", clean_root), ("lowlight", lowlight_root)]:
            class_dir = src_root / class_name
            assert class_dir.exists(), f"Missing class dir: {class_dir}"

            for src_path in sorted(class_dir.rglob("*")):
                if not src_path.is_file():
                    continue
                if src_path.suffix.lower() not in VALID_IMAGE_EXTENSIONS:
                    continue

                rel = src_path.relative_to(class_dir)
                rel_flat = "__".join(rel.parts)
                dst_name = f"{domain_name}__{rel_flat}"
                dst_path = dst_dir / dst_name

                shutil.copy2(src_path, dst_path)

                rows.append({
                    "split": split_name,
                    "class_name": class_name,
                    "domain": domain_name,
                    "source_path": str(src_path),
                    "dest_path": str(dst_path),
                })

copy_split_sources(TRAIN_CLEAN_ROOT, TRAIN_LOWLIGHT_ROOT, "train")
copy_split_sources(VAL_CLEAN_ROOT, VAL_LOWLIGHT_ROOT, "val")

with open(MIXED_MANIFEST_CSV, "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(
        f,
        fieldnames=["split", "class_name", "domain", "source_path", "dest_path"],
    )
    writer.writeheader()
    writer.writerows(rows)

print("Created mixed bundle:", MIXED_BUNDLE_ROOT)
print("Saved mixed manifest:", MIXED_MANIFEST_CSV)
print("Total copied rows:", len(rows))

```

```python
import torch
import numpy as np
import pprint

best_ckpt = torch.load(MIXED_BEST_CKPT, map_location="cpu", weights_only=False)

def to_python(obj):
    if isinstance(obj, dict):
        return {k: to_python(v) for k, v in obj.items()}
    if isinstance(obj, list):
        return [to_python(v) for v in obj]
    if isinstance(obj, tuple):
        return tuple(to_python(v) for v in obj)
    if isinstance(obj, np.ndarray):
        return obj.tolist()
    if isinstance(obj, np.generic):
        return obj.item()
    return obj

print("Checkpoint keys:")
print(sorted(best_ckpt.keys()))
print()

print("Epoch:", best_ckpt.get("epoch"))
print("Best threshold:", best_ckpt.get("best_threshold"))
print()

print("Train metrics:")
pprint.pprint(to_python(best_ckpt.get("train_metrics", {})))
print()

print("Mixed validation metrics:")
pprint.pprint(to_python(best_ckpt.get("val_metrics", {})))
print()

print("Low-light-only validation metrics:")
pprint.pprint(to_python(best_ckpt.get("lowlight_metrics", {})))
```
